Определим процент посещений сайта Яндекс.Метрики с бразузера Google Chrome

```python

browser_visits = pd.read_csv('jupyter_lessons/step_2_lesson_7.csv')

browser_visits
```

```python
visits_per_browser = browser_visits.groupby('browser') \
    .agg({'visits': 'sum'})

round(visits_per_browser.loc['Google Chrome'] / visits_per_browser.sum() * 100, 0)
```

Выясним, с какого браузера приходит больше всего ботов. Посмотрим на данные и определим, 
у какого браузера самая большая доля посещений сайта ботами от общего числа посещений

```python
visits_by_user = pd.read_csv('jupyter_lessons/step_3_lesson_7.csv')

visits_by_user
```

```python
# Создадим сводную таблицу
user_type = visits_by_user.groupby(['browser', 'user_type'], as_index=False) \
    .agg({'visits': 'sum'}) \
    .pivot(index='browser', columns='user_type', values='visits') \
    .fillna(0) 

user_type
```

```python
# Определим долю ботов
user_type.assign(bots_rate = user_type.Robots / (user_type.Robots + user_type.People)) \
    .sort_values('bots_rate') \
    .tail(1)
```

Сагрегируем данные по дню, типу пользователя, браузеру, создадим колонку с метрикой доли заходов на сайт для каждого из этих разрезов.
Вычислим процент открытий сайта настоящими пользователями Safari (не ботами)

```python
for_visits_rate = pd.read_csv('jupyter_lessons/step_4_lesson_7.csv')

for_visits_rate.head(7)
```

```python
visits_rate = for_visits_rate.assign(rate = round(100* (for_visits_rate.visits / for_visits_rate.visits.sum()), 1))

visits_rate
```
